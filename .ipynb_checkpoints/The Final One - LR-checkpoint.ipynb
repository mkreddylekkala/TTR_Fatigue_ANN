```python
import torch
import pandas as pd
import numpy as np
import torch.nn as nn # All neural network modules, nn.Linear, nn.Conv2d, BatchNorm, Loss functions
import torch.optim as optim # For all Optimization algorithms, SGD, Adam, etc.
import torch.nn.functional as F # All functions that don't have any parameters
from torch.utils.data import DataLoader, Dataset # Gives easier dataset managment and creates mini batches
from torch.utils.data import random_split
from scipy.stats import loguniform
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_size = 8 
output_size = 1
num_epochs = 2501
batchsize=64
L1=100
L2=75
```

```python
df=pd.read_csv('B1-Full.csv')
df.drop('MAPE',axis=1,inplace=True)
#df.drop('Case',axis=1,inplace=True)
df.drop(df.index[df['log(OMFD)'] == 0], inplace = True)
df.describe()
```

```python
max1=[0.761999756, 0.019099863, 2.719972336, 2.719487389, 14779.8172, 1499.980003, 1.0, 0.599912852]
min1=[0.152420269, 0.00700027, 0.000529749, 1.2e-05, 24.69063394, 500.0693348, 0.0, 0.300020114]
```

```python
df_z_scaled = df.drop('log(OMFD)',axis=1)
#apply normalization techniques
i=0
for col in df_z_scaled.columns:
    df_z_scaled[col] = (df_z_scaled[col]-min1[i])/(max1[i]-min1[i])
    i=i+1  
  
 #view normalized data   
display(df_z_scaled)
```

```python
class TTRDataset(Dataset):
    # load the dataset
    def __init__(self):
        
        self.X = torch.FloatTensor(df_z_scaled.values)
        self.y = torch.FloatTensor(np.log(10**df['log(OMFD)']).values)
        
  # number of rows in the dataset
    def __len__(self):
        return len(self.X)
 
    # get a row at an index
    def __getitem__(self, idx):
        return [self.X[idx], self.y[idx]]
 
    # get indexes for train and test rows
    def get_splits(self, n_valid=0.1, n_test=0.1):
        # determine sizes
        torch.manual_seed(0)
        valid_size = round(n_valid * len(self.X))
        test_size = round(n_test * len(self.X))
        train_size = len(self.X) - test_size - valid_size
        # calculate the split
        return random_split(self, [train_size, test_size, valid_size])

```

```python
    dataset=TTRDataset()
    train, valid, test = dataset.get_splits()
    train_loader = DataLoader(train, batch_size=batchsize, shuffle=True)
    valid_loader = DataLoader(valid,batch_size=batchsize,shuffle=False)
    test_loader = DataLoader(test,batch_size=batchsize,shuffle=False)
```

```python
class NN(nn.Module):
    def __init__(self, input_size, output_size):
        super(NN, self).__init__()
        self.fc1 = nn.Linear(input_size, L1)
        self.fc2 = nn.Linear(L1, L2)      
        self.out = nn.Linear(L2, output_size)
        
        #self.dropout = nn.Dropout(0.25)
        
        nn.init.kaiming_normal_(self.fc1.weight, mode='fan_in',nonlinearity='linear')
        nn.init.kaiming_normal_(self.fc2.weight, mode='fan_in',nonlinearity='linear')
        #nn.init.xavier_normal_(self.out.weight, gain=5/3)
        #nn.init.xavier_normal_(self.out.weight, gain=5/3)
        nn.init.xavier_normal_(self.out.weight, gain=1)
        
    def forward(self, x):
        x = F.elu((self.fc1(x)))
        #x = self.dropout(x)
        x = F.elu((self.fc2(x)))
        #x = self.dropout(x)
        x = self.out(x).reshape(-1)
        return x 
```

```python

rvs=[0.00075]



for j in (rvs):

    model = NN(input_size=input_size, output_size=output_size).to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=j, betas=(0.9, 0.999), eps=1e-08, weight_decay=1e-5, amsgrad=True)
    criterion = nn.L1Loss()
    #optimizer = torch.optim.SGD(model.parameters(), lr=rvs[j], momentum=0.9,weight_decay=1e-5)  
    
    train_losses1=[]
    valid_losses1=[]

    for epoch in range(num_epochs):
            train_losses=0
            valid_losses=0
            model.train()
            for batch_idx, (data, targets) in enumerate(train_loader):
                # Get data to cuda if possible
                data = data.to(device=device)
                targets = targets.to(device=device)

                # -1 will flatten all outer dimensions into one
                data = data.reshape(data.shape[0], -1) 

                # forward propagation
                scores = model(data)
                loss =criterion(scores, targets)
                train_losses+=loss.item()*data.size(0)
                # zero previous gradients
                optimizer.zero_grad()

                # back-propagation
                loss.backward()

                # gradient descent or adam step
                optimizer.step()

            train_losses1.append(train_losses/len(train_loader.dataset))

            model.eval()
            with torch.no_grad():
                for x, y in valid_loader:
                    x = x.to(device=device)
                    y = y.to(device=device)
                    x = x.reshape(x.shape[0], -1)

                    pred = model(x)
                    loss1=criterion(y,pred)
                    valid_losses+=loss1.item()*y.size(0)

                valid_losses1.append(valid_losses/len(valid_loader.dataset))
        
    #Saving model
            if(epoch%25==0):                
                FILE = 'models/LR_Test/MAE/new-eLU100-75/eLU100-75_lr'+str(j)+'MSE_checkpoint'+str(epoch)+'.pth'
                torch.save({
                            'epoch': epoch,
                            'model_state_dict': model.state_dict(),
                            'optimizer_state_dict': optimizer.state_dict(),
                            'loss_valid': valid_losses/len(valid_loader.dataset),
                            'loss_train': train_losses/len(train_loader.dataset),                                               
                           }, FILE)
                
    plt.plot(range(num_epochs),train_losses1,label='Train Loss')
    plt.plot(range(num_epochs),valid_losses1,label='CV Loss')
    plt.legend(loc="upper right")
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.show()

#Eval   

    def scores(model,loader):

            Pred=[]
            Y=[]
            model.eval()
            with torch.no_grad():
                for x, y in loader:
                    x = x.to(device=device)
                    y = y.to(device=device)
                    x = x.reshape(x.shape[0], -1)

                    pred = model(x)
                    Y.append(y.cpu().numpy())
                    Pred.append(pred.cpu().numpy())


            Y=np.concatenate(Y, axis=0)
            Pred=np.concatenate(Pred, axis=0)
            
            Y=np.exp(Y)
            Pred=np.exp(Pred)
            
            X1=np.abs((Y-Pred)/Y)*100
            p=0
            for i in X1:
                if i<=10:
                    p=p+1
            print('Below 10%:',p/len(loader.dataset))   
            
            p=0
            for i in X1:
                if i<=20:
                    p=p+1
            print('Below 20%:',p/len(loader.dataset))   
            
            
            MSE = np.mean((Y-Pred)**2)

            MAE = np.mean(np.abs(Y-Pred))

            MAPE = np.mean(np.abs((Y-Pred)/Y))*100

            R2 = 1-MSE/np.var(Y)

            if (len(loader.dataset)==len(train_loader.dataset)):
                print("Training Set -> lr: {}, R2: {:.3f}, MSE: {:.3f}, MAPE%: {:.3f}".format(j,R2,MSE,MAPE))
            else:
                print("Validation Set -> lr: {}, R2: {:.3f}, MSE: {:.3f}, MAPE%: {:.3f}\n".format(j,R2,MSE,MAPE))

    scores(model, train_loader)  
    scores(model, valid_loader)
```
